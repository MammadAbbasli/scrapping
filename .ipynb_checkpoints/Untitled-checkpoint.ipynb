```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
}
```

```python

requests.packages.urllib3.disable_warnings()
```

```python
prices = []
currencies = []
model = []
engine = []
name = []
code = []
year = []
km = []
city = []
time = []


for j in range(1,5):
    url = f'https://turbo.az/autos?page={j}&q%5Bavailability_status%5D=&q%5Bbarter%5D=0&q%5Bcrashed%5D=1&q%5Bcurrency%5D=azn&q%5Bengine_volume_from%5D=&q%5Bengine_volume_to%5D=&q%5Bfor_spare_parts%5D=0&q%5Bloan%5D=0&q%5Bmake%5D%5B%5D=4&q%5Bmileage_from%5D=&q%5Bmileage_to%5D=&q%5Bonly_shops%5D=&q%5Bpainted%5D=1&q%5Bpower_from%5D=&q%5Bpower_to%5D=&q%5Bprice_from%5D=&q%5Bprice_to%5D=&q%5Bregion%5D%5B%5D=&q%5Bsort%5D=&q%5Bused%5D=&q%5Byear_from%5D=&q%5Byear_to%5D='
    response = requests.get(url,headers=headers,verify = False)
    soup = BeautifulSoup(response.text,'html.parser')
    products = soup.find_all('div',class_='products-i')
    for i in products:
        bottom = i.find_all('div',class_='products-i__bottom')
        for bt in bottom:

            for pr in bt.find_all('div',class_="products-i__datetime"):
                pr_data = pr.text.split()

                city.append(pr_data[0])
                time.append(" ".join(pr_data[1:]))
            for pr in bt.find_all('div',class_="product-price"):
                pr_data = pr.text.split()
                prices.append(pr_data[0] + pr_data[1])
                currencies.append(pr_data[2])

            for pr in bt.find_all('div',class_="products-i__attributes products-i__bottom-text"):
                pr_data = pr.text.split()
                year.append(pr_data[0])
                engine.append(pr_data[1]+pr_data[2])
                km.append(" ".join(pr_data[3:]))


            for pr in bt.find_all('div',class_="products-i__name products-i__bottom-text"):
                names = pr.text.split()
                if len(names)>=3:
                    name.append(names[0])
                    code.append(' '.join(names[2:]))
                    model.append(names[1])
                else:
                    name.append(names[0])
                    code.append('nan')
                    model.append(names[1])
    print(f'{j}-ci seyfe')
```

```python
data = pd.DataFrame(model,columns=['Model'])
data["name"] = name 
data["prices"] = prices 
data["currencies"] = currencies 
data["engine"] = engine 

data["code"] = code 
data["year"] = year 
data["km"] = km 
data["city"] = city 
data["time"] = time 
```
