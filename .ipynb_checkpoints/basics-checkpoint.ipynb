```python
aa = soup.find_all("div",class_="products-i__bottom")
```

```python
for cc in aa:
    

    price_elements = soup.find_all('div', class_='product-price')

   
    for price_element in price_elements:
        data_list = price_element.text.strip().split()  
        price = data_list[0] + data_list[1]
        currency  = data_list[2]

        print(f"Price: {price}, Currency: {currency}")
```

```python
for cc in aa:
    names = cc.find_all('div',class_="products-i__bottom-text")
    for na in names:
        print(na)
```
